### Bivariate Analysis


#### BOX PLOT

```python
import seaborn as sns
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Fare by Passenger Class')
plt.show()
```

#### STACKED BAR CAHRT

```python
survival_counts = df.groupby('pclass')['survived'].value_counts().unstack()
colors = ['tab:blue', 'tab:green']
survival_counts.plot(kind='bar', stacked=True,color=colors)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Count')
plt.title('Survival Count by Passenger Class')
plt.show()
```

### Multivariate Analysis

#### SCATTER PLOT

```python
sns.scatterplot(x='age', y='fare', hue='pclass', data=df)
plt.title('Age vs. Fare (colored by pclass)')
plt.show()
```

#### CAT PLOT

```python
sns.catplot(x='pclass', hue='sex', col='survived', data=df, kind='count')
plt.suptitle('Survival by Passenger Class and Gender')
plt.show()
```

###  Find the outliers and replace the outliers

```python
import numpy as np
z_threshold = 3
z_scores = (df['age'] - df['age'].mean()) / df['age'].std()
outliers = (np.abs(z_scores) > z_threshold)
df.loc[outliers,'age'] = df['age'].mean()
```
